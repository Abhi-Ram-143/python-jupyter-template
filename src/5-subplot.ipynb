### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
bar_data = [5, 7, 3, 8, 6]
scatter_data_x = [1, 2, 3, 4, 5]
scatter_data_y = [10, 9, 7, 8, 6]
pie_data = [10, 20, 30, 40]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart in the first subplot
axs[0, 0].plot(x, y, label='Line Chart')
axs[0, 0].set_title('Line Chart')
axs[0, 0].legend()

# Bar chart in the second subplot
axs[0, 1].bar(x, bar_data, label='Bar Chart', color='orange')
axs[0, 1].set_title('Bar Chart')
axs[0, 1].legend()

# Scatter plot in the third subplot
axs[1, 0].scatter(scatter_data_x, scatter_data_y, label='Scatter Plot', color='green')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].legend()

# Pie chart in the fourth subplot
axs[1, 1].pie(pie_data, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
